```python
import  fastText
from collections import Counter
```

```python
import pandas as pd
```

```python
classifier = fastText.train_supervised('/home/tom/new_data/train_set.txt')


result = classifier.test('/home/tom/new_data/train_set.txt')
```

```python
classifier＝fastText
```

```python
print ('P@1:', result[0])
print ('R@1:', result[1])
print ('Number of examples:', result.nexamples)
```

```python
with open('/home/tom/new_data/train_set.txt') as file:
    lines=file.readlines()
result_id=[]
result_label=[]

for line in lines:
    (_id,text)=line.split(',')
    result_id.append(_id)
    result_label.append(classifier.predict(text.strip())[0][0].replace('__label__',''))
result={
    'id':result_id,
    'class':result_label
}

result_pd=pd.DataFrame.from_dict(result)
```

```python
result_pd.to_csv('/home/tom/new_data/result.csv',columns=['id','class'],index=False,index_label=False,header=False)
```

```python
with open('/home/tom/new_data/train_set.txt') as file:
    lines=file.readlines()
texts=[]
count=0
for line in lines:
    count+=1

    if count<20000:
        (text,_,_)=line.split('__')
        
        texts+=(text.split(' '))
train_set=Counter(texts)    
```

```python
with open('/home/tom/new_data/train_set_w.txt') as file:
    lines=file.readlines()
texts=[]
count=0
for line in lines:
    count+=1

    if count<20000:
        (text,_,_)=line.split('__')
        texts+=(text.split(' '))
train_set_w=Counter(texts) 


```

```python
roun1=[]
tsm=train_set.most_common(500000)
tsmw=train_set_w.most_common(500000)
for ts in tsm:
    _id,num=ts
    for tsw in tsmw:
        (_idw,num_w)=tsw
        if num==num_w:
            roun1.append(_id)
            break
        elif num>num_w:
            break
print(len(roun1))
```

```python
roun2=[]
tsm=train_set.most_common(600000)
tsmw=train_set_w.most_common(600000)
for ts in tsm:
    _id,num=ts
    for tsw in tsmw:
        (_idw,num_w)=tsw
        if num==num_w:
            if _id in roun1:
                roun2.append(_id)
            break
        elif num>num_w:
            break
print(len(roun2))
```
