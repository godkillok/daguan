```python
with open('/home/tom/new_data/train_set.txt') as file:
    lines=file.readlines()
texts=[]
count=0
for line in lines:
    count+=1

    if count<20000:
        (text,_,_)=line.split('__')
        
        texts+=(text.split(' '))
train_set=Counter(texts)    
```

```python
with open('/home/tom/new_data/train_set_w.txt') as file:
    lines=file.readlines()
texts=[]
count=0
for line in lines:
    count+=1

    if count<20000:
        (text,_,_)=line.split('__')
        texts+=(text.split(' '))
train_set_w=Counter(texts) 
```

```python
roun1=[]
tsm=train_set.most_common(500000)
tsmw=train_set_w.most_common(500000)
for ts in tsm:
    _id,num=ts
    for tsw in tsmw:
        (_idw,num_w)=tsw
        if num==num_w:
            roun1.append(_id)
            break
        elif num>num_w:
            break
print(len(roun1))
```

```python
roun2=[]
tsm=train_set.most_common(600000)
tsmw=train_set_w.most_common(600000)
for ts in tsm:
    _id,num=ts
    for tsw in tsmw:
        (_idw,num_w)=tsw
        if num==num_w:
            if _id in roun1:
                roun2.append(_id)
            break
        elif num>num_w:
            break
print(len(roun2))
```
